```
import tensorflow as tf
tf.__version__      # 런타임 재시작을 눌러서 버전을 변경한 후 버전 체크
```

```
a = 10
b = 20
ts = tf.add(a, b)   # tensorflow를 사용하여 코딩
print(ts)           # 30이라는 것은 나오질 않는다.
```

```
a = 10
b = 20
ts = tf.add(a, b)   

sess = tf.Session()   # tensorflow는 session을 이용하여 연산을 사용해야만 한다.
print(sess.run(ts))
sess.close()          # 항상 close 해줘야 한다.
```

```
x = 2
y = 3
pow1 = tf.pow(x, 2)
mul1 = tf.multiply(pow1, y)
add1 = tf.add(y, 2)
add2 = tf.add(mul1, add1)

with tf.Session() as sess:    # 이 구문을 사용하면 close를 안해줘도 된다.
  print(sess.run(pow1))
  print(sess.run(add2))
```

```
x = 2
y = 3
add_op = tf.add(x, y, name='Add')
mul_op = tf.multiply(x, y, name='Multiply')
pow_op = tf.pow(add_op, mul_op, name='Power')
useless_op = tf.multiply(x, add_op, name='Useless')

with tf.Session() as sess:
  print(sess.run(pow_op))
  print(sess.run(useless_op))
```

```
# 텐서플로 난수

r1 = tf.random_normal([1])
r2 = tf.random_normal([1])

print(r1)
print("===============")
print(r2)
print("===============")
with tf.Session() as sess:
  n1, n2 = sess.run([r1, r2])
  print(n1, n2)
```

```
# 텐서플로 난수 2

r1 = tf.random_normal([1, 3])   # 1행 3열 --> 즉 3개
r2 = tf.random_normal([1, 5])

print(r1)
print("==========")
print(r2)
print("==========")

with tf.Session() as sess:
  n1, n2 = sess.run([r1, r2])
  print(n1, n2)
```

```
# hello world 코딩  

hello = tf.constant('Hello World')
sess = tf.Session()

print(sess.run(hello))

sess.close()
```

```
# with 구문 example  

x = tf.constant(1.)
bool = tf.constant(True)
res = tf.cond(bool, lambda: tf.add(x, 1.), lambda: tf.add(x, 10.))

with tf.Session() as sess:
  print(sess.run(res))
```

```
# with 구문 example 2

a = tf.constant([5], dtype=tf.float32)
b = tf.constant([10], dtype=tf.float32)
c = tf.constant([2], dtype=tf.float32)

d = a * b + c

with tf.Session() as sess:
  print(sess.run(d))
```

```
# with 구문 example 3

a = tf.constant(5, dtype=tf.float32)
# a = 5

with tf.Session() as sess:
  print(sess.run(a))
```

```
# 텐서를 세션 없이 실행 - 세션을 가지고 실행

x = tf.constant([1, 2, 3])
y = tf.constant([5, 6, 7])

with tf.Session() as sess:
  print(sess.run(x + y))
```

```
# 텐서를 세션 없이 실행 - 세션을 가지고 실행 2  

x = tf.constant([1, 2, 3])
y = tf.constant([[1], [2], [3]])

with tf.Session() as sess:
  print(sess.run(x + y))
```

```
# 텐서를 세션 없이 실행 - 세션을 가지지 않고 실행 
# 그대로 출력 (그래프의 정보를 출력)

a = 2
b = 3 
c = tf.add(a, b, name='Add')
print(c)
```

```
# tf.reduce_mean example 

a = tf.constant(2)
b = tf.constant(3)
c = tf.constant(5)

# Some more operation
mean = tf.reduce_mean([a, b, c])
sum = tf.reduce_sum([a, b, c])

with tf.Session() as sess:
  print("mean = ", sess.run(mean))
  print("sum = ", sess.run(sum))
```

```
# 즉시 실행

tf.executing_eagerly()    # tensorflow 1.x 버전을 사용하고 있어서 False로 나온다. 즉, 즉시 실행이 기본이 아니다.
```

```
# 즉시 실행 예제

interactive_sess = tf.InteractiveSession()    # 1.x 버전은 즉시 실행을 이렇게 사용 (session 없이)
a = 2
b = 3
c = tf.add(a, b, name='Add')
print(c.eval())                               # 결과를 볼 때에는 메소드 eval() 사용
```

```
# 즉시 실행 예제 2

x = tf.constant(1.)
bool = tf.constant(True)
res = tf.cond(bool, lambda: tf.add(x, 1.), lambda: tf.add(x, 10.))

print(res.eval())
```
