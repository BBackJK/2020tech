```
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```
# 텐서플로 난수 - 균등분포 난수

# tf.random.uniform([1], 0, 1) - 배열, [시작, 끝)

rand = tf.random.uniform([1],0 ,1)
print(rand)

print("==============")

rand2 = tf.random.uniform([5, 4], 0, 1)
print(rand2)

print("==============")

rand3 = tf.random.uniform([1000], 0, 10)
print(rand3[:10])
```

```
# 균등 분포 1000개 그리기

rand4 = tf.random.uniform([10000], 0, 50)       # 1000개를 하면 10000개를 할 때 그래프보다 들쑥날쑥 하다 
plt.hist(rand4, bins=10)                        # 1000000개를 하면 더 평평함 (갯수가 많기 때문)
```

```
# 정규 분포 난수  

# tf.random.normal([4], 0, 1) - 크기, 평균, 표준편차  

rand5 = tf.random.normal([4], 0, 1)     # 3.9 랜덤한 수 여러개 얻기 
print(rand5)

print("================")

rand6 = tf.random.normal([2, 4],0, 2)   # 3.9 랜덤한 수 여러개 얻기
print(rand6)
```

```
# 정규 분포 100개 그리기

rand7 = tf.random.normal([100], 0, 1)
plt.hist(rand7, bins=10)
```

```
# 균등분포와 정규분포의 비교  

rand8 = tf.random.normal([1000], 0, 1)    # 정규 분포
rand9 = tf.random.uniform([2000], 0, 1)   # 균등 분포
                                          # 갯수를 더 많이 하면 들쑥날쑥한게 많이 사라진다

plt.hist(rand8, bins=100)
plt.hist(rand9, bins=100)
```

```
# 균등분포와 정규분포를 부분으로 그리기  

rand10 = tf.random.normal([1000], 0, 1)
rand11 = tf.random.uniform([2000], 0, 1)

plt.rcParams["figure.figsize"]= (12, 6)                     # 그림의 사이즈를 지정
fig, axes = plt.subplots(1, 2, sharex=True, sharey=True)    # 1행 2열
axes[0].hist(rand10, bins=100)                              # 1행 2열의 0 (왼쪽)
axes[1].hist(rand11, bins=100)                              # 1행 2열의 1 (오른쪽)
```
